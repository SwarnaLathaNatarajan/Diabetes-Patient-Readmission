```python
import numpy as np
import pandas as pd

# Visualization
import matplotlib.pyplot as plt
import seaborn as sns
import IPython.display as ipd
# import librosa.display
import plotly.offline as py
py.init_notebook_mode(connected=True)
import plotly.graph_objs as go
import plotly.tools as tls
import pandas as pd
```

```python
data = pd.read_csv("./diabetic_data.csv")
data.replace('?',np.nan,inplace=True)
data
```

```python
#Adding a new column "output label" which is 1 if readmitted
# data["output_label"]=(data.readmitted!='NO').astype('int')
data["output_label"]=(data["readmitted"].astype('str')).apply(lambda x: 0 if x == "NO" else (1 if x == "<30" else 2))
data
```

```python
numerical_cols=['time_in_hospital','num_lab_procedures','num_procedures',
                'num_medications','number_outpatient','number_emergency',
                'number_inpatient','number_diagnoses']
categorical_cols=[
        "gender",
        "age",
        "admission_type_id",
        "discharge_disposition_id",
        "admission_source_id",
        "medical_specialty",
        "number_outpatient",
        "number_emergency",
        "number_inpatient",
        "diag_1",
        "diag_2",
        "diag_3",
        "max_glu_serum",
        "A1Cresult",
        "metformin",
        "repaglinide",
        "nateglinide",
        "chlorpropamide",
        "glimepiride",
        "acetohexamide",
        "glipizide",
        "glyburide",
        "tolbutamide",
        "pioglitazone",
        "rosiglitazone",
        "acarbose",
        "miglitol",
        "troglitazone",
        "tolazamide",
        "examide",
        "citoglipton",
        "insulin",
        "glyburide-metformin",
        "glipizide-metformin",
        "metformin-rosiglitazone",
        "metformin-pioglitazone",
        "change",
        "diabetesMed"]
```

```python
data[numerical_cols].isnull().sum() #shows that it has no missing value
```

```python
data['race'].isnull().sum()#has null values
```

```python
data['race']=data['race'].fillna('UNK')
data['payer_code']=data['payer_code'].fillna('UNK')
data['medical_specialty']=data['medical_specialty'].fillna('UNK')
```

```python
data.groupby('medical_specialty').size().sort_values(ascending= False) #Shows that many unknown values exist
```

```python
selected_items=['UNK','InternalMedicine','Emergency/Trauma',
                'Family/GeneralPractice','Cardiology','Surgery-General',
               'Nephrology','Orthopedics','Orthopedics-Reconstructive',
               'Radiologist']
data['new_medical_specialty']=data['medical_specialty'].copy()
data.loc[~data.new_medical_specialty.isin(selected_items),'new_medical_specialty']='Other'
```

```python
data.groupby('new_medical_specialty').size()
```

```python
numerical_categorical_col=['admission_source_id','discharge_disposition_id','admission_type_id']
data[numerical_categorical_col]=data[numerical_categorical_col].astype('str')
```

```python
categorical_dataframe=pd.get_dummies(data[categorical_cols+numerical_categorical_col+['new_medical_specialty']], drop_first=True)
categorical_dataframe.head()
```

```python
data=pd.concat([data,categorical_dataframe],axis=1)
```

```python
list_of_cols_in_categorical_df=list(categorical_dataframe.columns)
```

```python
replace_age_dict={'[0-10)':0,'[10-20)':10,'[20-30)':20,
                  '[30-40)':30,'[40-50)':40,'[50-60)':50,'[60-70)':60,
                  '[70-80)':70,'[80-90)':80,'[90-100)':90}
data['age_group']=data.age.replace(replace_age_dict)
```

```python
data['haswt']=data.weight.notnull().astype('int')
```

```python
list_of_extracols=['age_group','haswt']
```

```python
final_cols=numerical_cols+list_of_cols_in_categorical_df+list_of_extracols
final_dataframe=data[final_cols]
```

```python
y=data['output_label']
print(final_dataframe.columns,y,sep="\n")
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import Pipeline,make_pipeline
from sklearn.model_selection import train_test_split
from sklearn import metrics
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.tree import DecisionTreeClassifier
X_train, X_test, y_train, y_test = train_test_split(final_dataframe,y)
models=[RandomForestClassifier()]#,DecisionTreeClassifier(),LogisticRegression(),GradientBoostingClassifier()]
for model in models:
    print(model)
    model.fit(X_train,y_train)
    predictions=model.predict(X_test)
    print(model.score(X_test,y_test))
    print(metrics.confusion_matrix(y_test, predictions))
```

```python
#INTERPRETABILITY - SHAP and LIME
import shap
import lime

explainerRF = shap.TreeExplainer(model[0]) #trying just for random forest
shap_values_RF_test = explainerRF.shap_values(X_test)
shap_values_RF_train = explainerRF.shap_values(X_train)
df_shap_RF_test = pd.DataFrame(shap_values_RF_test, columns=X_test.columns.values)
df_shap_RF_train = pd.DataFrame(shap_values_RF_train, columns=X_train.columns.values)
shap.initjs()
shap.force_plot(explainerRF.expected_value, shap_values_RF_test, X_test)
```
